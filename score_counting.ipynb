```python
import math
import numpy as np
import pandas as pd
```

```python
preDF = pd.read_csv('/home/ddl/Desktop/Notebooks/TBrain-submit/20180601_o_o.csv', thousands=',')
realDF = pd.read_csv('/home/ddl/Desktop/Notebooks/TBrain/20180608/18ETF.csv', thousands=',')
```

```python
preDF.sort_values(by='ETFid', ascending=True) 
```

```python
comArr = []
for i in range(5):
    a = realDF[realDF['日期'] == (20180604+i)].sort_values(by='代碼', ascending=True).iloc[:,6].tolist()
    print(a)
    b = preDF.sort_values(by='ETFid', ascending=True).iloc[:,2+i*2].tolist()
    print(b)
    comArr.append(b)
    comArr.append(a)   
```

```python
comDF = pd.DataFrame(comArr).T.rename(columns={0:'Mon_pred',1:'Mon_real',2:'Tue_pred',3:'Tue_real',4:'Wed_pred',5:'Wed_real',6:'Thus_pred',7:'Thus_real',8:'Fri_pred',9:'Fri_real'})
ans = []
for i in range(5):
    predict = comDF.iloc[:,i].tolist()
    real = comDF.iloc[:,i+1].tolist()
    p_score = [
        ( ( float(r) - abs(float(p) - float(r)) ) / float(r) )
        for (r, p) in zip(real, predict) 
    ]
    ans.append(predict)
    ans.append(real)
    ans.append(p_score)
ans
```

```python
ansDF = pd.DataFrame(ans).T.rename(columns={0:'Mon_pred',1:'Mon_real',2:'Mon_Sco',3:'Tue_pred',4:'Tue_real',5:'Tue_sco',6:'Wed_pred',7:'Wed_real',8:'Wed_sco',9:'Thus_pred',10:'Thus_real',11:'Thus_sco',12:'Fri_pred',13:'Fri_real',14:'Fri_sco'})
```
